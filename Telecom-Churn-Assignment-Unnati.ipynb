```python
## Read the data
import pandas as pd

dfChurn = pd.read_csv("telecom_churn_data.csv")
print(dfChurn.shape)
dfChurn.head()
```

```python
#Find the number of nulls in dataframe.
dfChurn.total_rech_amt_6.isnull().sum()
```

```python
#Find the number of nulls in dataframe.
dfChurn.total_rech_amt_6.isnull().sum()
```

```python
dfChurn["total_rech_amt_6_7"] = dfChurn[["total_rech_amt_6", "total_rech_amt_7"]].sum(axis=1)/2
dfChurn["total_rech_amt_6_7"]
```

```python
dfChurn = dfChurn.loc[dfChurn["total_rech_amt_6_7"] >= dfChurn["total_rech_amt_6_7"].quantile(0.7)]
dfChurn
```

```python
#Check for nulls in the columns used for churning
dfChurn.loc[:,['total_ic_mou_9', 'total_og_mou_9', 'vol_2g_mb_9','vol_3g_mb_9']].isnull().sum()
```

```python
dfChurn['churn'] = 0
dfChurn['churn']
```

```python
dfChurn.loc[((dfChurn.total_ic_mou_9 == 0) & (dfChurn.total_og_mou_9 == 0) & (dfChurn.vol_2g_mb_9 ==0) & (dfChurn.vol_3g_mb_9 ==0)),'churn']=1
dfChurn['churn']
```

```python
# drop all columns corresponding to the attributes of churn phase

dfChurn.drop(dfChurn.columns[dfChurn.columns.str.contains('_9')],axis=1,inplace=True)
dfChurn.shape
```

```python
ls_nochange = ['circle_id','loc_og_t2o_mou','std_og_t2o_mou','loc_ic_t2o_mou','last_date_of_month_6','last_date_of_month_7','last_date_of_month_8','std_og_t2c_mou_6','std_og_t2c_mou_7','std_og_t2c_mou_8','std_ic_t2o_mou_6','std_ic_t2o_mou_7','std_ic_t2o_mou_8']
```

```python
#Lets have a look at average revenue per user

avgRevUser = dfChurn.columns[dfChurn.columns.str.contains('arpu',regex=True)]

pd.DataFrame([dfChurn.loc[dfChurn.churn==0,avgRevUser].mean()
              ,dfChurn.loc[dfChurn.churn==1,avgRevUser].mean()],index={'Non Churned','Churned'})
```
